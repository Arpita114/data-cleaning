# Data Cleaning & Preprocessing Automation Tool - Analysis Notebook

This notebook demonstrates the complete workflow:
- Load
- Inspect
- Clean
- Validate
- Export


```python
import sys
sys.path.append('../src')
import pandas as pd
from loader import DataLoader
from inspector import DataInspector
from cleaner import DataCleaner
from validator import DataValidator
from exporter import DataExporter

# Load
df = DataLoader.load_csv('../data/raw_data.csv')
df.head()
```

```python
# Inspect
inspector = DataInspector(df)
inspector.display_info()
inspector.check_missing_values()
inspector.check_duplicates()
inspector.validate_dtypes()
```

```python
# Clean
cleaner = DataCleaner(df)
cleaned_df = cleaner.full_cleaning_pipeline(missing_strategy='auto')
cleaned_df.head(10)
cleaner.get_cleaning_log()
```

```python
# Visualize missing values (optional)
try:
    import matplotlib.pyplot as plt
    df.isnull().sum().plot(kind='bar', title='Missing Values Before Cleaning')
    plt.show()
    cleaned_df.isnull().sum().plot(kind='bar', title='Missing Values After Cleaning')
    plt.show()
except Exception as e:
    print(e)
```

```python
# Validate
validator = DataValidator(cleaned_df)
validation = validator.validate()
validation
```

```python
# Export & Report
exporter = DataExporter()
exporter.export_csv(cleaned_df, filename='../data/cleaned_data.csv')
exporter.generate_cleaning_report(
    original_info=inspector.get_basic_info(),
    cleaning_log=cleaner.get_cleaning_log(),
    final_df=cleaned_df,
    validation_result=validation,
    report_path='../reports/cleaning_report.txt'
)
```
